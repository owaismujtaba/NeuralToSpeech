```python
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt
import os
import seaborn as sns
import pandas as pd
```

```python
predicted_neuro = np.load('NeuroIncept/predicted.npy')
predicted_FCN = np.load('FCN/predicted.npy')
predicted_CNN = np.load('CNN/predicted.npy')
original = np.load('NeuroIncept/original.npy')
```

```python
predicted_neuro.shape
```

```python
original.shape
```

```python
def plot_spectrograms(start, end):
    fig, ax = plt.subplots(2, 2, figsize=(12, 6), sharex=True, sharey=True)

    # Plot original spectrogram
    im = ax[0, 0].imshow(original[start:end].T, aspect='auto', origin='lower', cmap='viridis')
    ax[0, 0].set_title('Original Log Mel Spectrogram', fontsize=16, fontweight='bold')

    # Plot predicted spectrogram (NeuroIncept Decoder)
    ax[0, 1].imshow(predicted_neuro[start:end].T, aspect='auto', origin='lower', cmap='viridis')
    ax[0, 1].set_title('NeuroIncept Decoder')

    # Plot predicted spectrogram (CNN)
    ax[1, 0].imshow(predicted_CNN[start:end].T, aspect='auto', origin='lower', cmap='viridis')
    ax[1, 0].set_title('CNN')

    # Plot predicted spectrogram (FCN)
    ax[1, 1].imshow(predicted_FCN[start:end].T, aspect='auto', origin='lower', cmap='viridis')
    ax[1, 1].set_title('FCN')

    # Only show x and y ticks on the first plot
    for i in range(2):
        for j in range(2):
                ax[i, j].tick_params(labelleft=False, labelbottom=False)

    plt.tight_layout()
    plt.savefig('spectrogram_comparison.png', dpi=600)
    plt.show()
```

```python

plot_spectrograms(3000, 3500)
```
