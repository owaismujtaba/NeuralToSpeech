```python
folder = 'correlations'
files = os.listdir(folder)
files = [Path(folder, item) for item in files if item.endswith('.npy') and item.startswith('sub')]
files
```

```python
names = []
for item in files:
    name = str(item).split('/')[-1].split('.')[0]
    names.append(name)
    data = pd.read_csv(item)
    plt.plot(data['val_loss'])
plt.legend(names)
plt.tight_layout()
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.savefig('losses.png', dpi=600)
    
```

```python
full_data = []
names = []
means = []
stds = []
for item in files:
    name = str(item).split('/')[-1].split('.')[0]
    names.append(name)
    data = np.load(item)
    full_data.append(data)
    means.append(np.mean(data))
    stds.append(np.std(data))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace with your actual data)
# full_data = [np.random.normal(size=100) for _ in range(5)]
# names = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']

# Calculate means and standard deviations
means = [np.mean(group) for group in full_data]
std_devs = [np.std(group) for group in full_data]

# Sort the means, std_devs, and names according to increasing means
sorted_indices = np.argsort(means)
sorted_means = [means[i] for i in sorted_indices]
sorted_std_devs = [std_devs[i] for i in sorted_indices]
sorted_names = [names[i] for i in sorted_indices]

plt.figure(figsize=(12, 7))

# Create the bar plot with error bars for sorted data
plt.bar(np.arange(len(sorted_means)), sorted_means, yerr=sorted_std_devs, capsize=10, color='lightblue', edgecolor='black')

# Set y-axis limit from 0.5 onwards
plt.ylim(bottom=0.5)

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add titles and labels
plt.xlabel('Participants', fontsize=14, fontweight='bold')
plt.ylabel('Correlations', fontsize=14, fontweight='bold')
plt.xticks(ticks=np.arange(len(sorted_names)), labels=sorted_names, rotation=45, ha='right', fontsize=12)

plt.tight_layout()  # Adjust layout to make sure everything fits well

# Save the plot as a high-resolution image
plt.savefig('sorted_correlations.png', dpi=600)

# Show the plot
plt.show()
```
